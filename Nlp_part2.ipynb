```python
from gensim.models import Word2Vec
sentences=[
    ["i","love","coding"],
    ["i","love","python"],
    ["python","is","powerful"]
]
#train CBOW model(sg=0)

cbow_model=Word2Vec(sentences,
                    min_count=1,
                    vector_size=5,#size of word vector
                    window=2,     #content window
                    sg=0)        #0-cbow
#get word vector
print("vector for coding")
print(cbow_model.wv["coding"])

#find similar words
print("similar to coding")
print(cbow_model.wv.most_similar("coding"))


```

```python
import nltk
nltk.download('all')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
stop_words = set(stopwords.words('english'))
le=WordNetLemmatizer()

def Lowercasing(word) :
  word=word.lower()
  return word

def Tokenization(word) :
  return word_tokenize(word)

def StopwordRem(toks):
  return [tok for tok in toks if tok not in stop_words]

def RemPuncuations(words) :
  return [word for word in words if word not in string.punctuation]

def Lemmatization(words) :
  return [le.lemmatize(word) for word in words]


for words in df['cols']:
  word=Lowercasing(words)
  tok=Tokenization(word)
  v2=StopwordRem(tok)
  v3=RemPuncuations(v2)
  v4=Lemmatization(v3)
  print(v4)






```

```python
import spacy
from collections import Counter
import pandas as pd

nlp = spacy.load("en_core_web_sm")
```

```python
def preprocess_spacy(text):
    doc = nlp(text.lower())  # Lowercase + tokenize
    tokens = [token.lemma_ for token in doc
              if not token.is_stop and token.is_alpha]  # remove stopwords & punctuation
    return tokens

# Apply to dataframe column
df['processed'] = df['cols'].apply(preprocess_spacy)
print(df[['cols','processed']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Join tokens back to string for TF-IDF
df['processed_text'] = df['processed'].apply(lambda x: " ".join(x))

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df['processed_text'])

print("TF-IDF Feature size:", X_tfidf.shape)
```
