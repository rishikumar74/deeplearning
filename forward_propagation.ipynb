```python
import numpy as np
x1=np.array([0,0,1,1])
x2=np.array([0,1,0,1])
y=np.array([0,1,1,0])
```

```python
w1=1
w2=1
w3=1
w4=1
w5=1
w6=-2
epochs=10
b1=-0.5
b2=-1.5
b3=-0.5

def activation(z) :
  return 1 if z>=0 else 0

for epoch in range(epochs) :
  print(f"epoch {epoch+1}")
  for i in range(len(x1)) :
    z1=w1*x1[i]+w2*x2[i]+b1
    ac1=activation(z1)
    z2=w3*x1[i]+w4*x2[i]+b2
    ac2=activation(z2)
    op=w5*ac1+w6*ac2+b3
    y_pred=activation(op)
    print(f"input x1:{x1[i]} x2:{x2[i]} actual op:{y[i]} y_pred:{y_pred}")

```
